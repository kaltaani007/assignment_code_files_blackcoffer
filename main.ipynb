## Getting all the titles of the website links ....

```python
all_headers = []
def links_process1(url):
    
    # using the BeautifulSoup module
    r = requests.get(url)
    if r.status_code == 200 :
        soup = BeautifulSoup (r.content , 'html.parser')
        all_headers.append(soup.title.get_text())
        
    else :
        all_headers.append("None")
```

# Find all the paragraphs 

```python
# creating a handler function to save info 

cont = []

def save_data (url ):
    
    r = requests.get(url)
    if r.status_code == 200:
        
        soup = BeautifulSoup(r.content , "html.parser")
        p_tree = soup.find("div" , class_ = "td-post-content")
        
        #l_text = []
        text = ""
        #print(type(p_tree))
        pts = p_tree.find_all('p')
        
        for x in pts:
            
            text += x.get_text()
            
        cont.append(text)
    
         
    else:
        cont.append("No Content ")
    
    
```

```python
for l in links:
    save_data(l)
```

```python
#len(cont)
```

```python
df2["web-content"] = cont 
df2.head()
```

### Functions

```python
def avg_words(x):
    words = x.split()
    return sum(len(word) for word in words) / len (words)
```

```python
def avg_sen_len(x):
    words = x.split(".")
    return sum(len(word) for word in words) / len (words)
```

```python
def count_syllables(word):
    
    return float(len(
        re.findall('(?!e$)[aeiouy]+', word, re.I) +
        re.findall('^[^aeiouy]*e$', word, re.I)
    )/len(word))

#count_syllables("test")
```

```python
pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
pronouns = pronounRegex.findall(' I me  you We and all of us make this team tweek, he is from US')
#print(len(pronouns))
```

```python
def pronoun_counter(x):
    pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    pronouns = pronounRegex.findall(x)
    return len(pronouns)
```

```python
def syllable_counter(x):
    words = x.split()
    return sum(count_syllables(word) for word in words) / len (words)
```

```python
df3['SYLLABLE_PER_WORD'] = df3["web-content"].apply(lambda x: syllable_counter(x) )
df3["AVERAGE_SENTENCE_LENGTH"] = df3["web-content"].apply(lambda x: avg_sen_len(x))
df3["AVERAGE_WORDS"] = df3["web-content"].apply(lambda x: avg_words(x))
df3['PERSONAL_PRONOUNS'] = df3["web-content"].apply(lambda x: pronoun_counter(x) )
```

```python
## Finding out the complex words 
### Removing simple words result out having complex words
```

```python
#!pip install nltk 
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
#len(stop_words)
```

```python
df3['COMPLEX_WORDS'] = df3["web-content"].apply(lambda x: len([word for word in x.split() if word.lower() not in stop_words])) 
df3['PER_COMPLEX_WORDS'] = df3["web-content"].apply(lambda x: (len([word for word in x.split() if word.lower() not in stop_words]))%100 )
```

```python
from textblob import TextBlob
```

```python
df3["POLARITY_SCORE"] = df3["web-content"].apply(lambda x: TextBlob(x).sentiment[0])
df3["SUBJECTIVITY_SCORE"] = df3["web-content"].apply(lambda x: TextBlob(x).sentiment[1])
#df.head()
```

```python
def positive_score(text):
    score = 0 
    
    score = TextBlob(text).sentiment[0]
    
    if score > 0 :
        return score
    else:
        return 0 
```

```python
def negative_score(text):
    score = 0 
    
    score = TextBlob(text).sentiment[0]
    
    if score < 0 :
        return score
    else:
        return 0 
```

```python
df3["NEGATIVE_SCORE"] = df3["web-content"].apply(lambda x: negative_score(x))
df3["POSITIVE_SCORE"] = df3["web-content"].apply(lambda x: positive_score(x))
```

```python
# Calculate the fog index 
#fog = 0.4 * ((word_count / len(sentence)) )

df3["FOG_INDEX"] = 0.4* (( df3['WORD_COUNT'] / df3['SENTENCE_COUNT']) + df3['COMPLEX_WORDS'])
```
